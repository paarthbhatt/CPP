```python
def pattern1(n):
    for i in range(n):
        for j in range(n):
            print("* ", end="")
        print()
pattern1(5)
```

```python
def pattern2(n):
    for i in range(n):
        for j in range(i + 1):
            print("* ", end="")
        print()

pattern2(5)
```

```python
def pattern3(n):
    for i in range(1, n + 1):
        for j in range(1, i + 1):
            print(j, end=" ")
        print()

pattern3(5)
```

```python
def pattern4(n):
    for i in range(1, n + 1):
        for j in range(1, i + 1):
            print(i, end=" ")
        print()

pattern4(5)
```

```python
def pattern5(n):
    for i in range(n):
        for j in range(n, i, -1):
            print("*", end=" ")
        print()

pattern5(5)
```

```python
def pattern6(n):
    for i in range(n):
        for j in range(n, i, -1):
            print(n - j + 1, end=" ")
        print()

pattern6(5)
```

```python
def pattern7(n):
    for i in range(n):
        for j in range(n-i-1):
            print(" ", end=" ")
        for k in range(2*i+1):
            print("*", end=" ")
        for l in range(n-i-1):
            print(" ", end=" ")
        print()

pattern7(5)
```
